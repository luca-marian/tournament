#Mapare commercial name cu naics label

```python
business_naics_path = '/content/drive/My Drive/HBN/business_taxonomy_labeled_mpnet.csv'
taxonomy = pd.DataFrame(pd.read_csv(business_naics_path))
taxonomy
```

```python

# Iterate through the merged DataFrame
naics_labels = []

for index, row in data.iterrows():
    # Access data in each row
    category = row['main_business_category']
    naics_labels.append(taxonomy[taxonomy['label'] == category]['naics_label'].values[0])

naics_labels
```

```python
data['naics_label'] = naics_labels
data
```

```python
sen_b = business_taxonomy['description']
sen_naics = naics_taxonomy['description']
from sentence_transformers import SentenceTransformer
#model = SentenceTransformer('bert-base-nli-mean-tokens')
model = SentenceTransformer('all-mpnet-base-v2')

sen_embeddings_b = model.encode(sen_b)
sen_embeddings_b.shape

sen_embeddings_naics = model.encode(sen_naics)
sen_embeddings_naics.shape
```

```python
sen_embeddings_b.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

results = cosine_similarity(
    [sen_embeddings_b[0]],
    sen_embeddings_naics[0:]
)
results
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_values = []
label = []

for i in range(0,len(sen_embeddings_b)):
  results = cosine_similarity(
      [sen_embeddings_b[i]],
      sen_embeddings_naics[0:]
  )
  similarity_values.append(results.max())
  label.append(naics_taxonomy.iloc[results.argmax()]['naics_label'])

business_taxonomy.loc[:, "naics_label"] = label
business_taxonomy
```
