```python
# Iterate through the merged DataFrame
data = pd.read_parquet('/content/drive/My Drive/HBN/tournament_hints_data.parquet', engine='pyarrow')

naics_labels = []

for index, row in data.iterrows():
    # Access data in each row
    category = row['main_business_category']
    naics_labels.append(taxonomy[taxonomy['label'] == category]['naics_label'].values[0])

data
```

```python
data['naics_label'] = naics_labels
data
```

```python
unique_rows = data.drop_duplicates(subset=['naics_label'], keep='last', ignore_index=True)
unique_rows
```

```python
sen = unique_rows['business_tags']
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')

sen_embeddings = model.encode(sen)
sen_embeddings.shape

np.save(output_path + 'embeddings_business_tags_94.npy', sen_embeddings)
```
